## DistilBERT base cased distilled SQuAD
DistilBERT is a small, fast, cheap and light Transformer model trained by distilling BERT base. More details [here](https://huggingface.co/distilbert/distilbert-base-cased-distilled-squad).

```python
model_name = "distilbert/distilbert-base-cased-distilled-squad"
```

```python
# TODO: Create a tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
text = "Big black bug bleeds black blood"
# pt = PyTorch tensor
enc_text = tokenizer(text, return_tensors='pt')
print(enc_text)
```

```python
dec_text = tokenizer.decode(enc_text.input_ids[0])
print(dec_text)
```

```python
# don't print the special tokens
dec_text = tokenizer.decode(enc_text.input_ids[0], skip_special_tokens=True)
print(dec_text)
```

```python
# TODO: Encode text
text = "Big black bug bleeds black blood"
text = "Thou shalt not make a machine in the likeness of the mind."

# pt = PyTorch tensors
enc_text = tokenizer(text, return_tensors='pt')

print(enc_text)
print('----------------------')

for k in enc_text:
  print(k, enc_text[k])
```

```python
for t in enc_text.input_ids[0]:
  dec_token = tokenizer.decode(t)
  print(t, dec_token)
```

```python
# TODO: Encoding multiple texts
texts = [
    "Big black bug bleeds black blood",
    "Thou shalt not make a machine in the likeness of the mind."
]

tok_texts = tokenizer(texts, return_tensors='pt', padding=True)

print(tok_texts)
```

## Working with LLMs
Create and instance of the Large Language Model (LLM). We will then create a simple
prompt, tokenize the prompt and feed the tokenized prompt to the LLM. The response
from the LLM will be decoded to human friendly text.

```python
# TODO: Load libraries
from transformers import AutoModelForQuestionAnswering
```

```python
# The model must be the same as the tokenizer!!!
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
# TODO: Load question answer model
idx = 10
context = dataset['train'][idx]['story_structure']
#context = dataset['train'][idx]['infilled_story']
question = dataset['train'][idx]['question']
expected_answer = dataset['train'][idx]['expected_answer']
```

```python
# TODO: Encode context and question
# order is important
enc_question = tokenizer(question, context, return_tensors='pt', padding=True)
print(enc_question)
```

```python
ans =
print(tokenizer.decode(enc_question.input_ids[0]))
```

```python
# TODO: Question the model
# input_ids, attention_mask
result = model(enc_question.input_ids, enc_question.attention_mask)

```

```python
print(result)
```

```python
import torch

# Find the start of the answer phrase
ans_start = torch.argmax(result.start_logits)
# Find the end of the answer phrase
ans_end = torch.argmax(result.end_logits)
print(f'start = {ans_start}, end = {ans_end}')
print(f'value = {result.start_logits[ans_start]}, value = {result.end_logits[ans_end]}')

# Extract the tokens
ans_ids = enc_question.input_ids[0][ans_start: ans_end + 1]

print(ans_ids)
```

```python
# Convert ans_ids to text
answer = tokenizer.decode(ans_ids)

print(answer)
```
